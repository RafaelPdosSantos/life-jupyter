## lançamento Projéctil
$ v_L = \frac{2 R^2 (\rho_{oleo} - \rho_{ar}) g} {9 \eta_{ar}} $
$ x = x_0 + v_0x t$ 

$ y = t_0 + v_{0_y} t - \frac{1}{1} g t^2 $ 

```python
#%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
#from matplotlib.ticker import MultipleLocator
import numpy as np
from scipy import constants # Constantes Físicas
from scipy import stats
import ipywidgets as widgets

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
#import ipywidgets as widgets

```

```python
t = np.linspace(0, 3.0, 101)
#angulo de lancamento
v0=10.0 # m/s
tetha=80/180.0*np.pi 
np.sin(tetha)
g=9.8
```

```python
x=v0*np.cos(tetha)*t
y=v0*np.sin(tetha)*t - 0.5*g*t*t
#y=- 0.5*g*t*t
```

```python
%matplotlib notebookplt.plot(x, y, 'r', label='fitted line')
#plt.legend()
plt.axis('equal')
plt.grid('on')
#    plt.tick_params
#   plt.plot(y, z, label='Cicloide')
plt.axis([0., 10, 0, 10])   # [ymin, ymax, zmin, zmax]
plt.show()
```

```python
%matplotlib notebook
t = np.linspace(0, 3.0, 101)
def plotBalistic(B):
    tetha = 80/180.0*np.pi
    x=v0*np.cos(tetha)*t
    y=v0*np.sin(tetha)*t - 0.5*g*t*t

    plt.plot(x, y)
    plt.axis('equal')
    plt.grid('on')
    plt.tick_params
    #plt.plot(y, z, label='Cicloide')
    plt.axis([0., 10, 0, 10])   # [ymin, ymax, zmin, zmax]
#    plt.legend(loc='upper center')
    plt.show()
    
widgets.interact(plotBalistic, B=(0.0,90.0,2.0));
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))

def update(w = 1.0):
    line.set_ydata(np.sin(w * x))
    fig.canvas.draw()

interact(update);
```
