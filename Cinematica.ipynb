```python
%matplotlib notebook
v0=10.0 # m/s
#angulo de lancamento
tetha=80/180.0*np.pi 
np.sin(tetha)
g=9.8
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
t = np.linspace(0, 2.0, 50)
x=v0*np.cos(tetha)*t
y=v0*np.sin(tetha)*t - 0.5*g*t*t
ax.axis('equal')
ax.grid('on')
#ax.legend(loc='upper center')

ax.axis([0., 14.0, 0.0, 4.0])
line, = ax.plot(x, y, 'r', label='Trajectoria')
ax.legend(loc='upper center')

def update(theta = (0.0,90.0,5.0)):
    thet = theta/180.0*np.pi
    x=v0*np.cos(thet)*t
    y=v0*np.sin(thet)*t - 0.5*g*t*t
    line.set_ydata(y)
    line.set_xdata(x)
    fig.canvas.draw()

interact(update);
```
